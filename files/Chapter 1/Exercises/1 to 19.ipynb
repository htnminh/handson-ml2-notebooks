# 17. What is the purpose of a validation set?


Validation set:
- Purpose: *Compare models:*
    - Select model
    - Tune hyperparameters
- Usage:
    - For every model:
        - Train: (reduced) training set (excluding validation set)
        - Validate: validation set
    - Select: the best validated model:
        - Train: (full) training set (including validation set)
        - Test: test set 


# 18. What is the train-dev set, when do you need it, and how do you use it?
Train-dev set:
- When to use: data mismatch between
    - training set: nonrepresentative
    - validation set and test set: representative
- Definition:
    - Splitted from: training set (excluding validation set)
    - To conclude if: overfit or data mismatch
- How to use:
    - Train: training set (excluding validation set and train-dev set)
    - *If model perform well on training set,* evaluate on train-dev set:
        - poor: overfit
        - well: not overfit
            - Validate on validation set: 
                - poor: data mismatch: try to preprocessing data, then retrain the model

# 19. What can go wrong if you tune hyperparameters using the test set?

The generalization error:
- measured multiple times
- on test set

$\Rightarrow$ The best measured model:
- adapted to that particular set
- $\Rightarrow$ might not perform well: new data

*(risk of overfitting, in short)*
